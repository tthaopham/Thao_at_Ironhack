```python
from stackapi import StackAPI
```

```python
SITE = StackAPI('stackoverflow')

# Q1 - questions & answers last month
answers_lm = SITE.fetch('answers', fromdate=1672012800, todate=1674691200)
answers_lm

questions_lm = SITE.fetch('answers', fromdate=1672012800, todate=1674691200)
questions_lm

# alternative options - use datetime library from python to use functions like date(), today(), etc
```

```python
# Q2 - most voted question:
questions = SITE.fetch('questions', fromdate=1674209408, todate=1674727808,
                        min=5, tagged='python', sort='votes', filter = 'withbody')
questions
```

```python
# Q3 - Find answer with id 6784, 6473
answers_id = SITE.fetch('answers', filter = "withbody" ,ids = [6784, 6473])
answers_id
```
