The steps to solve this problem is similar to the previous one except that you need to find out the repository names instead of developer names.

````python
# This is the url you will scrape in this exercise
url_git = 'https://github.com/trending/python?since=daily'
response_git = requests.get(url_git)
html_git = response_git.content
soup_git = bs4.BeautifulSoup(html_git, "html.parser")


soup_site = [element.text.strip().replace("\n","").replace(" ","") for element in soup_git.find_all('h1', class_ = 'h3 lh-condensed')]
#soup_site = list of elements (not the text itself --> cannot use function list)
````

````python
# your code here
soup_h1_list = [element for element in soup_git.find_all('h1', class_ = 'h3 lh-condensed')]
soup_a_list = [element.find('a')['href'] for element in soup_h1_list]
soup_a_list
````

#### Display all the image links from Walt Disney wikipedia page.

````python
# This is the url you will scrape in this exercise
url_disney = 'https://en.wikipedia.org/wiki/Walt_Disney'
response_disney = requests.get(url_disney).content
response_disney

soup_disney = bs4.BeautifulSoup(response_disney, "html.parser")
# print(soup_disney) ==> this works

image_url_list = soup_disney.find_all('a', class_ = 'image')
# find_all will result into a list, so it is not an element

image_url_list_a = [a.find('img')['src'] for a in image_url_list]
image_url_list_a

````

#### Find the number of titles that have changed in the United States Code since its last release point.

````python
# This is the url you will scrape in this exercise
url_cd = 'http://uscode.house.gov/download/download.shtml'
res_cd = requests.get(url_cd)
soup_cd = bs4.BeautifulSoup(res_cd.content, 'lxml')
titles = [title.text.strip() for title in soup_cd.find_all('div', class_ = 'usctitlechanged')]
titles
````
